# Structural type of dwelling and household size: Canada, provinces and territories, census metropolitan areas and census agglomerations with parts

```python
struct_dwel_df = pd.read_csv('../input/structural-type-of-dwelling-household-size-canada/98100040.csv')
struct_dwel_df.head()
```

```python
struct_dwel_df.shape
```

```python
struct_dwel_df.info()
```

```python
# Finding number of unique values in each column

struct_dwel_df.nunique(dropna=True).sort_values()
```

Total household size should be enough. We dont want granular data for 1 person, 2 person, etc. So we can drop all the columns for 1,2,3,4,and 5+ persons and its corresponding index which has all null values

```python
# struct_dwel_df.drop(struct_dwel_df.columns[[6,16]], axis=1, inplace=True)
struct_dwel_df.drop(struct_dwel_df.iloc[:, 6:16], inplace = True, axis = 1)
```

```python
struct_dwel_df.info()
```

```python
struct_dwel_df.drop(struct_dwel_df.columns[[6, 8, 10]], axis=1, inplace=True)
```

```python
struct_dwel_df.info()
```

```python
#renaming the columns for readability
struct_dwel_df.rename(columns = {'Structural type of dwelling (9)':'Structural type of dwelling', 
                                 'Household size (8):Total - Household size[1]':'Total - Household',
                                 'Household size (8):Number of persons in private households[7]':'Number of persons in private households', 
                                 'Household size (8):Average household size[8]' : 'Average household size'},
                                 inplace = True)
```

```python
struct_dwel_df.head()
```

**The above table has Structural type of dwelling and household size: Canada, provinces and territories, census metropolitan areas and census agglomerations with parts**

```python
struct_dwel_df.to_csv('Structural-dwelling-household-size-edited.csv',index=False)
```

# Population and dwelling counts: Canada and population centres 

```python
population_df = pd.read_csv('../input/Population-dwelling-counts/98100011.csv')
population_df.head()
```

```python
population_df.info()
```

Remove all the columns with null values (symbols column)

```python
population_df.drop(population_df.columns[[5, 7, 9, 11, 13, 15, 17]], axis=1, inplace=True)
```

```python
population_df.info()
```

```python
#renaming the columns for readability
population_df.rename(columns = {'Population and dwelling counts (7): Population, 2021 [1]':'Population,2021', 
                                 'Population and dwelling counts (7): Population, 2016 [2]':'Population,2016',
                                 'Population and dwelling counts (7): Population percentage change, 2016 to 2021 [3]':'Population percentage change', 
                                 'Population and dwelling counts (7): Total private dwellings, 2021 [4]' : 'Total private dwellings,2021',
                                 'Population and dwelling counts (7): Private dwellings occupied by usual residents, 2021 [5]':'Private dwellings occupied by usual residents, 2021', 
                                 'Population and dwelling counts (7): Land area in square kilometres, 2021 [6]' : 'Land area in square kilometres,2021',
                                 'Population and dwelling counts (7): Population density per square kilometre, 2021 [7]' : 'Population density per square kilometre,2021',
                                    
                                },
                                 inplace = True)
```

```python
population_df.head()
```

```python
population_df.to_csv('population_df_edited.csv',index=False)
```
